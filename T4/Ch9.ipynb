# Inheritance
- a class can inherit from other class
- it improve code reuse
- construcor ,attribute and method are inheritance to child class
- parent class has no access to child class
- child class methods overrides the parent class methods
 - this is called method overriding 
- super() method is used to invoke parent class method and constructor

## `super() method`
- if we write constructor in both parent and child class,the parent class constrc is not available in child class
- `super()` method is used to call parent class from the child class

```python
class Phone:
    def __init__(self,price,brand,camera):
        print("inside phone class")
        self.price=price
        self.brand=brand
        self.camera=camera
    def buy(self):
        print("Buying phone")
class SmartPhone(Phone):
    def buy(self):
        print("Buying a smartphone")
        super().buy()
```

```python
s=SmartPhone(20000,'iphone',13)
```

## Abstract class
- a class derived from BAC class which belong to abc module is called abstract class in python
- ABC class is known as meta class which means class that defines behaviour of other classes.
- Abstarct class can have concreate and abstract methods
- Abstract class needs to be extended and its methods implemented 
- one cannot create object of an abstract class

```python
from abc import ABC,abstractmethod
class DefenceForce(ABC):
    def __init__(self):
        self.id=101
    @abstractmethod
    def area(self):
        pass
    def gun(self):
        print("the gun is ak47")
```

```python
class Army(DefenceForce):
    def area(self):
        print("area is land",self.id)
```

```python
a=Army()
```

```python
a.area()
```

```python
a.gun()
```

```python
class Navy(DefenceForce):
    def area(self):
        print("area is sea",self.id)
```

```python
n=Navy()
```

```python
n.area()
```

```python
n.gun()
```

```python
n.id
```

## Create an abstract class Shape with abstract method `calculate_area`. Create two class Rectangle  and  Circle inheriting from shape. Create two object of rectangle and circle and find the area of these objects

```python
from abc import ABC,abstractmethod
import math
class Shape(ABC):
    @abstractmethod
    def calculate_area():
        pass
    
```

```python
class Rectangle(Shape):
    def __init__(self,l,b):
        self.l=l
        self.b=b
    def calculate_area(self):
        print(self.l*self.b)
```

```python
class Circle(Shape):
    def __init__(self,r):
        self.r=r
    def calculate_area(self):
        print(math.pi*(self.r**2))
```

```python
r=Rectangle(10,20)
r.calculate_area()
```

```python
c=Circle(10)
c.calculate_area()
```

## Method resolution order in python
- It defines an order in which the class is searched  in case of multiple inheritance
- It is based on C3 Linearlization Algorith (>=python2.3)

## C3 linearization algorithm
- For a list of classes - C1C2C3..Cn
  - the C1 is the head of class and the rest is tail
- C3 linearization algorith is defined as MRO of sum of any class C and linear of parent and the list of parents

    L(C(B1,B2...BN)) = C + merge (L(B1),L(B2)...BN))
- Here B1 to BN ae the parent classes of C
1. Take the head of first list L(B1) from left to right
2. If the head of selected class is not in the tail of the list add it to the linear and remove it from the list 
3. Repeat the process untill the classes are removed or it is impossible to find good head

```python
class F:
    pass
class E:
    pass
class D:
    pass
class C(D,F):
    pass
class B(D,E):
    pass
class A(B,C):
    pass
```

```python
A.mro()
```

### First step
- MRO OF D,E,F,O
- L(D) = DO
- L(E) = EO
- L(F) = FO
- L(O) = O


### Second step
- L(C)=C+MERGE(L(D),L(F),DF)
-     =C+ MERGE(DO,FO,DF)
-     =C+D+MERGE(O,OF,F)
-     =C+D+F+O
-     =CDFO

### Third Step
- L(B) = B + merge(L(D),L(E),DE)
-     = B + MERGE(DO, EO, DE)
-     = B + D + MERGE(O, EO, E)
-     = B + D + E + O
-     = BDEO


## Four Step
- L(A) = A + MERGE(L(B),L(C),BC)
-     = A + MERGE(BDEO,CDFO,BC)
-     = A + B + MERGE(DEO , CDFO , C)
-     = A + B + C+ MERGE(DEO,DFO)
-      = A + B + C+D MERGE(EO,FO)
-      = A + B + C+ D + E +F + O
    

```python
class A:  #AO
    pass
class B:      #BO
    pass
class C:      #CO
    pass
class D:     #DO
    pass
class E:     #EO
    pass
class K1(C,A,B):      # K1CABO
    pass
class K3(A,D):      #K3AD
    pass
class K2(B,D,E):      #K2BDE
    pass
class Z( K1,K3,K2):      #Z + MERGE(K1CABO,K3AD,K2BDE+ZK1K3K2)
                        # ZK1CK3AK2BDEO
                        # ZK1CK3AK2BDEO
            
    pass
```

### OPERATOR OVERLOADING
__add__= +
__sub__= +
__mul__= +
__truediv__= +

```python
class Fraction:
    def __init__(self,x,y):
        self.num=x
        self.den=y
    def __add__(self,other):
        new_num=self.num*other.den + other.num*self.den
        new_den=self.den * other.den
        return f"{new_num}/{new_den}"
    def __sub__(self,other):
        new_num=self.num*other.den - other.num*self.den
        new_den=self.den * other.den
        return f"{new_num}/{new_den}"
    def __mul__(self,other):
        new_num=self.num*other.num
        new_den=self.den * other.den
        return f"{new_num}/{new_den}"
    def __truediv__(self,other):
        new_num=self.num*other.den
        new_den=self.den * other.num
        return f"{new_num}/{new_den}"
```
