## for a given list of integer A , let S  be sum of digit of minimum element of A; return 0 if S is odd otherwise 1
- [34,33,1,24,75,33,54,8] output 0 
- [99,77,33,66,55]   output 1

````python
def sumdigit(A):
    mini=min(A)
    result=0
    while mini>0:
        quo=mini%10
        rem=mini//10
        result+=quo
        mini=rem
    return 1 if result%2==0 else 0
````

````python
l=[34,33,1,24,75,33,54,8]
sumdigit(l)
````

````python
def findchar(s,char):
    for i in range(len(s)):
        if s[i]==char:
            return i
    return -1
````
