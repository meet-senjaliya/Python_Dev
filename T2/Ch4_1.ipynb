```python
s="this is a python tutorial"
space=0
word=0
for i in s:
    if i==' ':
        space+=1
    else:
        word+=1
print(space)
print(word)
```

## maketrans and translate() method
- These methods are used to replace individual characters from a given string

```python
intab="aeiou"
outtab="12345"
mystring = "This is a string function"
table = mystring.maketrans(intab,outtab)
print(table)
newstring =  mystring.translate(table)
print(newstring)
```

## Remove punctuation

```python
import string
string.punctuation
```

```python
mystring = "T&h%is i!s [a] st^rin?g f{unctio}n"
table = mystring.maketrans('i','$',string.punctuation)
print(table)
newstring =  mystring.translate(table)
print(newstring)
```
