```python
s1='ljku'
s2='lk'
a=len(s1)
b=len(s2)
valid=True
if a>b:
    for i in s2:
        if s2.find(i)==-1:
            valid=False
    if valid:
        print("yes")
    else:
        print("false")
if a<b:
    for i in s1:
        if s2.find(i)==-1:
            valid=False
    if valid:
        print("yes")
    else:
        print("false")
```

## WAP to return max even number from a given tuple
- input (8,7,2,3,4,5,100,3,56)
- output 100

```python
def maximum1(t):
    max1=0
    for i in t:
        if i%2==0 and i>max1:
            max1=i
    print(max1)
```

```python
t=(1,2,8,100,101)
maximum1(t)
```

## WAP to replace special symbol in a given string with '#'

```python
str1="abc5@a#c$e"
str2 =''
for i in str1:
    if ((ord(i)>=65 and ord(i)<=90) or (ord(i)>=97) and (ord(i)<=122) or (ord(i)>=49) and (ord(i)<=57)):
        i=i
    else:
        i='#'
    str2=str2+i
print(str2)
```
